```python
import numpy as np
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt

import matplotlib.collections as mcol
from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple
from matplotlib.lines import Line2D

import random 
```

```python
random.seed(400)

y = np.random.random(10)
x = np.arange(y.size)
```

```python
new_len = 50
new_x = np.linspace(x.min(), x.max(), new_len)
new_y = interpolate.interp1d(x,y, kind='cubic')(new_x)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x, y, 'bo-')
plt.title("Amostra original")
```

```python
plt.subplot(2,1,1)
plt.plot(new_x, new_y, 'ro-')
plt.title("Amostra com interpolação")
```

```python
plt.style.use('_mpl-gallery')

x = np.linspace(-10,10,100)
y = 4*np.sin(2 * x)

fig, ax = plt.subplots()

ax.plot(x,y,linewidth = 2.0)

ax.set(xlim =(-5,5), xticks=np.arange(-5,5), ylim =(-5,5), yticks=np.arange(-5,5))

plt.show
```

Exemplo de gráfico Scatter com legenda

```python
t1 = np.arange(0.0,2.0, 0.1)
t2 = np.arange(0.0,2.0, 0.01)

fig, ax = plt.subplots()

l1 = ax.plot(t2, np.exp(-t2))
l2, l3 = ax.plot(t2, np.sin(2 * np.pi * t2), '--o', t1, np.log(1 + t1), '.')
l4, = ax.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 's-.')

ax.legend((l2, l4), ("Legenda 1", "Legenda 2"), loc = 'upper right', shadow = True)

plt.show()
```

```python
#Curva predeterminada
#Perceba que o própio python realiza a aproximação por pontos. Dado um conjunto x de pontos,
#podemos realizar uma aproximação de curva.

x = np.linspace(-5,5,5) #Terceira coordena que determina a quantidade de pontos
y = x**(3) + 2*x**(2) - 16

fig, ax = plt.subplots()
ax.plot(x,y,linewidth = 1)

```
